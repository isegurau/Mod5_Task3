```python
#replace values for string category
credit["SEX"]=credit["SEX"].replace({1: "Male", 2: "Female"})

credit["EDUCATION"]=credit["EDUCATION"].replace({1: "school", 2: "university", 3:"high school", 4:"Other"})

credit["MARRIAGE"]=credit["MARRIAGE"].replace({1: "married", 2: "single", 3:"divorce", 4:"Other"})


credit["PAY_0"]=credit["PAY_0"].replace({-2: "No Consumption", -1: "Paid Full", 0:"Use Revolving", 1:"Delayed 1 month",2:"Delayed 2 months", 3:"Delayed 3 months"
                                        ,4:"Delayed 4 month",5:"Delayed 5 months", 6 :"Delayed 6 months", 7:"Delayed 7 months", 8:"Delayed 8 months", 9:"Delayed 9 or above"})

print(credit)

```

```python
#parse columns to category
credit['SEX'] = credit.SEX.astype('category')
credit['EDUCATION'] = credit.EDUCATION.astype('category')
credit['MARRIAGE'] = credit.MARRIAGE.astype('category')
```

```python
credit.describe()
```

```python
#Create a new column for the age category
bins = [20, 35, 50, 65, np.inf]
names = ['20-35', '35-50', '50-65', '65+']

credit['AgeRange'] = pd.cut(credit['AGE'], bins, labels=names)

print(credit.dtypes)
```

```python
#delete unnecesary columns
credit=credit.drop(columns=['ID', 'default payment next month', 'AGE'])
```

```python
#delete columns by CORRELATION >90
credit= credit.drop(columns=['BILL_AMT4', 'BILL_AMT5','BILL_AMT6'])
```
